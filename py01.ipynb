```python
# filter 는 true인 값만 남김
data = [1, -2, 3, 4, -5, 0, -1, 3, -4, 5]

def pos(v):
    return v > 0  # 양수인 경우 True, 음수인 경우 False를 반환

print(list(filter(pos, data)  ))
```

```python
lis = list(range(1, 11))  # 1부터 10까지의 숫자를 리스트로 생성


print(list(filter(odd, lis)))  # 홀수만 필터링
print(list(filter(even, lis)))  # 짝수만 필터링
print(list(map(add, lis)))  # 각 요소에 10을 더함
print(list(map(mul, lis)))  # 각 요소에 10을 곱함      
print(list(map(addmul, lis)))  # 각 요소에 10을 더한 후 곱함

def odd(x):
    return x % 2 == 1  # 홀수인 경우 True, 짝수인 경우 False를 반환
def even(x):
    return x % 2 == 0  # 짝수인 경우 True, 홀수인 경우 False를 반환
def addmul(x):
    xx = add(x)  # 10을 더한 후
    xxx = mul(xx)  # 그 결과에 10을 곱함
    return (xxx) * x # 10을 더한 후 곱하는 함수

def add(x):
    return x + 10   # 10을 더하는 함수 
def mul(x):
    return x * 10 # 10을 곱하는 함수
```

```python
# 예외처리

student_score = [90,82,45,71,60, -200, 55, 100, 99, 88, 77, 110]

f = open("C:\\phj\\student_scores2.txt", "w", encoding="utf-8")
try:
    for score in student_score:
        if score < 0 or score > 100:
            f.write(f"에러발생 중단")
            raise ValueError("점수는 0에서 100 사이여야 합니다.")
        
        if score >= 90 and score <= 100:
            grade = "A"
        elif score >= 80 and score < 90:
            grade = "B"
        elif score >= 70 and score < 80:
            grade = "C" 
        elif score >= 60 and score < 70:
            grade = "D"
        else:
            grade = "F"
        f.write(f"점수: {score}, 등급: {grade}\n")  # 각 학생의 점수와 등급을 파일에 저장
    print("파일로 정상 저장 되었습니다.")
except Exception as e:
    print(f"오류 발생: {e}")  # 예외 발생 시 오류 메시지 출력
    print(f"※파일 비정상 저장※")
finally:
    f.close()  # 파일을 닫음
    print("파일이 닫혔습니다.")  # 파일 닫힘 메시지 출력
```

```python
# datetime 모듈
import datetime
day1 = datetime.date(2025, 1, 22)  # 날짜 객체 생성
day2 = datetime.date(2025, 8, 14)  # 또 다른 날짜 객체 생성
diff = day2 - day1  # 날짜 차이 계산
diff2 = day2 - datetime.date.today()  # 현재 날짜와 day2의 차이 계산
print(f"{day1}와 {day2}의 차이는 {diff.days}일입니다.")  # 날짜 차이 출력
print("="*40)
print("전체 수업기간:", diff.days, "일")  # 전체 수업기간 출력
print("남은 수업기간:", diff2.days, "일")  # 남은 수업기간 출력
print("지난 수업기간:", diff.days - diff2.days, "일")  # 지난 수업기간 출력 
```

```python
import random
lucky_combo = []  # 로또 번호를 저장할 리스트

while True:
    lucky_num = random.randint(1, 45)  # 1부터 45까지의 랜덤 정수 생성
    
    if len(lucky_combo) == 6:  # 로또 번호가 6개가 되면 반복문 종료
        break

    if lucky_num not in lucky_combo:
        lucky_combo.append(lucky_num)
    
lucky_combo.sort()  # 로또 번호를 오름차순으로 정렬
print("당첨 번호:", lucky_combo)  # 로또 번호 출력
    
```

```python
import itertools# 순열과 조합
students = ['철수', '영희', '민수', '지영']
snacks = ['과자', '사탕', '초콜릿'] 
result = zip(students, snacks)  # 두 리스트를 묶어서 튜플로 반환
print(list(result))  # 결과를 리스트로 변환하여 출력
```

```python
import itertools# 순열과 조합
students = ['철수', '영희', '민수', '지영','정수']
snacks = ['과자', '사탕', '초콜릿', ] 
result = zip(students, snacks)  # 두 리스트를 묶어서 튜플로 반환
print(list(result))  # 결과를 리스트로 변환하여 출력

result2 = itertools.zip_longest(students, snacks, fillvalue='칸초')  # 두 리스트를 묶어서 튜플로 반환, 짧은 리스트는 '없음'으로 채움
print(list(result2))  # 결과를 리스트로 변환하여 출력
```
